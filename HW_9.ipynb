## Задание.

В этом задании нужно убедиться, что ЦПТ действительно работает.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n и постройте гистрограммы средних этих выборок.

```python
# Параметры запуска: размеры выборок и количество повторов генерации
rv = sts.norm()

fig, ax = plt.subplots(1, 1)

ax.figure.set_size_inches(25, 20)

ax.set_ylim([0, 15])
ax.set_xlim([-0.5, 0.5])

for n in range(100,1000,47):
    ma = []
    for number in range(1000):
        # На каждом цикле генерируем выборки разных размеров
        sample = rv.rvs(size=n)
        # Считаем среднее по каждой выборке и добавляем в массив средних
        ma.append(sample.mean())
        label_1 = str(n)
    ax.hist(ma, density=True, label=label_1)
plt.title('NORMAL')
ax.legend(loc='best', frameon=False)
plt.show()
```

```python
# Проверим на экспоненциальном распределении

rv = sts.expon()

fig, ax = plt.subplots(1, 1)

ax.figure.set_size_inches(25, 20)

ax.set_ylim([0, 15])
ax.set_xlim([0, 2])

for n in range(100,1000,53):
    ma = []
    for number in range(10000):
        # На каждом цикле генерируем выборки разных размеров
        sample = rv.rvs(size=n)
        # Считаем среднее по каждой выборке и добавляем в массив средних
        ma.append(sample.mean())
        label_1 = str(n)
    ax.hist(ma, density=True, label=label_1)
plt.title('NORMAL')
ax.legend(loc='best', frameon=False)
plt.show()
```

## Вывод

ЦПТ работает: На примере двух начальных выборок разных распределений итоговая выборка из средних приближается к нормальному распределению и тем ближе, чем большее число испытаний проведено.
