## Уровень 0:

- Задание 1:


Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.


```python
fig, ax = plt.subplots(1, 1)

ax.figure.set_size_inches(25, 10)
ax.set_ylim([0, 1])
ax.set_xlim([0, 25])

test_coef = [[10, 1], [20, 0.5], [20, 3], [5, 3], [15, 4]]
for k in test_coef:

    n = norm(loc=k[0], scale = k[1])
    n.rvs(size=500)

    x = np.linspace(0,100,10000)

    pdf = n.pdf(x)

    ax.plot(x, pdf, label=f'coef {k[0]},{k[1]}')
plt.title('NORMAL')
ax.legend(loc='best', frameon=False)
plt.show()
```

```python
fig, axe = plt.subplots(1, 1)

axe.figure.set_size_inches(25, 10)
axe.set_ylim([0, 3])
axe.set_xlim([4, 35])

test_coef = [[10, 1], [20, 0.5], [20, 3], [5, 3], [15, 4]]
for k in test_coef:

    n = expon(loc=k[0], scale = k[1])
    n.rvs(size=500)

    x = np.linspace(0,100,10000)

    pdf = n.pdf(x)

    axe.plot(x, pdf, label=f'coef {k[0]},{k[1]}')
plt.title('EXPON')
axe.legend(loc='best', frameon=False)
plt.show()
```

```python
fig, axe = plt.subplots(1, 1)

axe.figure.set_size_inches(25, 10)
axe.set_ylim([0, 0.5])
axe.set_xlim([-10, 10])

test_coef = [0.01, 0.3, 1, 5, 50000]
for k in test_coef:

    n = stu(k)
    n.rvs(size=25)

    x = np.linspace(-15,15,5000)

    pdf = n.pdf(x)

    axe.plot(x, pdf, label=f'freedoms {k}')
plt.title('STUDENT')
axe.legend(loc='best', frameon=False)
plt.show()
```

- Задание 3


Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

```python
full_circle = 54+4+28
redp = 54/full_circle
yellowp = 4/full_circle
greenp = 28/full_circle

# величины неравномерно распределены и поэтому формула Шеннона

I = -1*(redp*math.log(redp, 2)+yellowp*math.log(yellowp, 2)+greenp*math.log(greenp, 2))
print(f'При варианте 1 (неравномерном горении светофора) получаем {I}')

# При равномерном распределении (все горят по 28с) будет применима формула Хартли
p = 1/3
n = 1/p
I = -1*(math.log(p,2))
check = ( -1*(math.log(p,2)) == math.log(n,2))

print(f'При варианте 2 (равномерном горении светофора) получаем {I}. Проверка выдает: {check}')
```
